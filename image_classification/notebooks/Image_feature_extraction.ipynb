```python
# from this guy: https://stackoverflow.com/questions/44793221/python-fractal-box-count-fractal-dimension
def rgb2gray(rgb):
    r, g, b = rgb[:,:,0]/255, rgb[:,:,1]/255, rgb[:,:,2]/255
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray



# From https://github.com/rougier/numpy-100 (#87)
def boxcount(Z, k):
    S = np.add.reduceat(
        np.add.reduceat(Z, np.arange(0, Z.shape[0], k), axis=0),
                        np.arange(0, Z.shape[1], k), axis=1)

    # We count non-empty (0) and non-full boxes (k*k)
    return len(np.where((S > 0) & (S < k*k))[0])
    
def fractal_dimension(Z, threshold=0.75):
    # Only for 2d image
    #assert(len(Z.shape) == 2)

    # Transform Z into a binary array
    Z = (Z < threshold)

    # Minimal dimension of image
    p = min(Z.shape)

    # Greatest power of 2 less than or equal to p
    n = 2**np.floor(np.log(p)/np.log(2))

    # Extract the exponent
    n = int(np.log(n)/np.log(2))

    # Build successive box sizes (from 2**n down to 2**1)
    sizes = 2**np.arange(n, 1, -1)

    # Actual box counting with decreasing size
    counts = []
    for size in sizes:
        counts.append(_boxcount(Z, size))

    # Fit the successive log(sizes) with log (counts)
    coeffs = np.polyfit(np.log(sizes), np.log(counts), 1)
    return -coeffs[0]
```

```python
def get_crop(img, plot=False):
    lbp =  img.copy()
    rv = img[:,:,0].ravel()
    gv = img[:,:,1].ravel()
    bv = img[:,:,2].ravel()

    rv_med = np.quantile(rv, q=0.33)
    gv_med = np.quantile(gv, q=0.33)
    bv_med = np.quantile(bv, q=0.33)

    rv_med_med = np.quantile(rv[rv<rv_med], q=0.5)
    gv_med_med = np.quantile(gv[gv<gv_med], q=0.5)
    bv_med_med = np.quantile(bv[bv<bv_med], q=0.5)

    lbp[lbp[:,:,0]>rv_med_med]=1
    lbp[lbp[:,:,1]>gv_med_med]=1
    lbp[lbp[:,:,2]>bv_med_med]=1
    lbp[lbp[:,:,0]<rv_med_med]=0
    lbp[lbp[:,:,1]<gv_med_med]=0
    lbp[lbp[:,:,2]<bv_med_med]=0

    flbp =(lbp[:,:,0]+lbp[:,:,1]+lbp[:,:,2])/3

    crop = 0.85
    impow = 1
    cropL = crop*np.min([width, height])

    cL = int(width/2)-int(0.5*cropL)
    cR = int(width/2)+int(0.5*cropL)
    cD = int(width/2)-int(0.5*cropL)
    cU = int(width/2)+int(0.5*cropL)
    flbp = flbp[cL:cR, cD:cU]

    sumb = np.sum(flbp==0)
    rad = int(1.75*np.sqrt(sumb/np.pi))
    #rad = int(0.75*rad) if rad>np.min(flbp.shape) else rad

    cx, cy = sc.ndimage.center_of_mass(1-flbp)
    cx = int(cx)
    cy = int(cy)+int(0.5*rad)

    cxl, cxr = np.max([cx-rad, 0]), np.min([cx+rad, flbp.shape[0]])
    cyd, cyu = np.max([cy-rad, 0]), np.min([cy+rad, flbp.shape[1]])

    intsx = np.sum(flbp, axis=0)
    intsy = np.sum(flbp, axis=1)

    flbp[cxl:cxr, -5+cyd:cyd+5] = 0
    flbp[cxl:cxr, -5+cyu:cyu+5] = 0
    flbp[-5+cxl:cxl+5, cyd:cyu] = 0
    flbp[-5+cxr:cxr+5, cyd:cyu] = 0

    if plot:
        fig, ax = plt.subplots(ncols=2, nrows=2, figsize=(14,8))
        ax[0,0].imshow(flbp[:, :], cmap=plt.cm.gray)
        ax[1,0].plot(np.arange(0,flbp.shape[0]), intsx)
        ax[0,1].plot(intsy, np.arange(flbp.shape[0], 0, -1))
        ax[0,1].set_yticks(np.arange(0, flbp.shape[0], 50))
        ax[0,1].set_yticklabels(np.arange(0, flbp.shape[0], 50)[::-1])
        ax[1,1].plot(np.sort(intsx), np.sort(intsy))
    
    return cxl, cxr, cyd, cyu
```

```python
@jit
def convolve2D(img):
    _ker = [[-1, -1.41,    -1], 
            [-1.41,  8, -1.41], 
            [-1, -1.41,   -1]]
    kernel = np.array([_ker, _ker, _ker])
    
    xr = np.arange(1, img.shape[0]-1)
    yr = np.arange(1, img.shape[1]-1)
    
    cimg = np.zeros(img.shape)
    
    for m in range(img.shape[2]):
        for k in xr:
            for l in yr:
                cimg[k, l, m] = np.sum(np.multiply(img[k-1:k+2, l-1:l+2, m], kernel[:,:,m]))
    return cimg
```

```python
def get_features(img, filename):
    features = ['filename', 'width', 'height', 'cx', 'cy', 'elongation',
                'r_diff_x', 'r_diff_y', 'g_diff_x', 'g_diff_y', 'b_diff_x', 'b_diff_y',
                'num_centers_aff', 'center_diff',
                'r_median', 'g_median', 'b_median',
                'r_std', 'g_std', 'b_std', 'r_autocorr', 'g_autocorr', 'b_autocorr',
                'rg_corr', 'rb_corr', 'gb_corr', 'rg_corr_var', 'rb_corr_var', 'gb_corr_var',
                'red_skew', 'blue_skew', 'green_skew',
                'rg_dot', 'rb_dot', 'bg_dot', 'fractal_dim_transform',
                'mis_RG', 'mis_RB', 'mis_GB', 'cos_RG', 'cos_RB', 'cos_GB'] 
    results = namedtuple('ImageFeatures', " ".join(features))
    
    width, height = img.shape[0], img.shape[1]

    crop = 0.9
    cropL = crop*np.min([width, height])
    cL = int(width/2)-int(0.5*cropL)
    cR = int(width/2)+int(0.5*cropL)
    cD = int(height/2)-int(0.5*cropL)
    cU = int(height/2)+int(0.5*cropL)
    img = img[cL:cR, cD:cU,:]

    img = skimage.exposure.equalize_adapthist(img)
    
    #img_transform = np.std(img[:,:,:], axis=2) 

    cy, cx = ndi.center_of_mass(skimage.color.rgb2gray(img))/cropL    
    
    try:
        _contours = skimage.measure.find_contours(skimage.color.rgb2gray(img), .5)
        _contour = max(_contours, key=len)
        elongation = (_contour[:,0].max()-_contour[:,0].min())/(_contour[:,1].max()-_contour[:,1].min())
    except Exception as e:
        elongation = 0
        print('Did not find contours, elongation set to zero')

    r_diff_x = np.mean(np.diff(img[:,:,0], axis=0))/255
    r_diff_y = np.mean(np.diff(img[:,:,0], axis=1))/255

    g_diff_x = np.mean(np.diff(img[:,:,1], axis=0))/255
    g_diff_y = np.mean(np.diff(img[:,:,1], axis=1))/255

    b_diff_x = np.mean(np.diff(img[:,:,2], axis=0))/255
    b_diff_y = np.mean(np.diff(img[:,:,2], axis=1))/255
    
    cl_affprop = AffinityPropagation(affinity='precomputed', max_iter=250, random_state=7)
    cl_affprop.fit(-skimage.color.rgb2gray(img))
    centers = cl_affprop.cluster_centers_indices_
    num_centers_aff = centers.shape[0] if isinstance(centers, np.ndarray) else 0
    if num_centers_aff>1:
        center_diff = (np.max(cl_affprop.cluster_centers_indices_)-np.min(cl_affprop.cluster_centers_indices_))/cropL
    else:
        center_diff = 0 
    
       
    r_median = np.median(img[:,:,0])
    g_median = np.median(img[:,:,1])
    b_median = np.median(img[:,:,2])

    r_std = np.std(img[:,:,0])
    g_std = np.std(img[:,:,1])
    b_std = np.std(img[:,:,2])

    r_autocorr = np.mean(np.corrcoef(img[:,:,0]))
    g_autocorr = np.mean(np.corrcoef(img[:,:,1]))
    b_autocorr = np.mean(np.corrcoef(img[:,:,2]))

    rg_corr = np.mean(np.corrcoef(img[:,:,0], img[:,:,1]))
    rb_corr = np.mean(np.corrcoef(img[:,:,0], img[:,:,2]))
    gb_corr = np.mean(np.corrcoef(img[:,:,1], img[:,:,2]))

    rg_corr_var = np.var(np.corrcoef(img[:,:,0], img[:,:,1]))
    rb_corr_var = np.var(np.corrcoef(img[:,:,0], img[:,:,2]))
    gb_corr_var = np.var(np.corrcoef(img[:,:,1], img[:,:,2]))

    red_vector = img[:,:,0].ravel()
    green_vector = img[:,:,1].ravel()
    blue_vector = img[:,:,2].ravel()
    
    red_skew = stats.skew(red_vector)
    green_skew = stats.skew(green_vector)
    blue_skew = stats.skew(blue_vector)
        
    rg_dot = np.dot(red_vector, green_vector)
    rb_dot = np.dot(red_vector, blue_vector)
    bg_dot = np.dot(blue_vector, green_vector)
    
    mis_RG = v_measure_score(red_vector, green_vector)
    mis_RB = v_measure_score(red_vector, blue_vector)
    mis_GB = v_measure_score(green_vector, blue_vector)
    
    cos_RG = sc.spatial.distance.cosine(red_vector, green_vector)
    cos_RB = sc.spatial.distance.cosine(red_vector, blue_vector)
    cos_GB = sc.spatial.distance.cosine(green_vector, blue_vector)
    
    qRu = np.quantile(red_vector, q=0.15)
    qRd = np.quantile(red_vector, q=0.025)
    qGu = np.quantile(green_vector, q=0.15)
    qGd = np.quantile(green_vector, q=0.025)
    qBu = np.quantile(blue_vector, q=0.15)
    qBd = np.quantile(blue_vector, q=0.025)
   
    canny1 = skimage.feature.canny(img[:,:,0], low_threshold=qRd, high_threshold=qRu)
    canny2 = skimage.feature.canny(img[:,:,1], low_threshold=qGd, high_threshold=qGu)
    canny3 = skimage.feature.canny(img[:,:,2], low_threshold=qBd, high_threshold=qBu) 
    
    fractal_dim_transform = fractal_dimension(canny1+canny2+canny3)
    
    return  results(filename=filename, width=width, height=height, cx=cx, cy=cy,
                    elongation=elongation, r_diff_x=r_diff_x, r_diff_y=r_diff_y,
                    g_diff_x=g_diff_x, g_diff_y=g_diff_y, b_diff_x=b_diff_x, 
                    b_diff_y=b_diff_y, num_centers_aff=num_centers_aff, center_diff=center_diff,
                    r_median=r_median, g_median=g_median, b_median=b_median,
                    r_std=r_std, g_std=g_std, b_std=b_std, 
                    r_autocorr=r_autocorr, g_autocorr=g_autocorr, b_autocorr=b_autocorr, 
                    rg_corr=rg_corr, rb_corr=rb_corr, gb_corr=gb_corr,
                    rg_corr_var=rg_corr_var, rb_corr_var=rb_corr_var, gb_corr_var=gb_corr_var,
                    red_skew=red_skew, green_skew=green_skew, blue_skew=blue_skew, 
                    rg_dot=rg_dot, rb_dot=rb_dot, bg_dot=bg_dot,
                    fractal_dim_transform=fractal_dim_transform, 
                    mis_RG=mis_RG, mis_RB=mis_RB, mis_GB=mis_GB,
                    cos_RG=cos_RG, cos_RB=cos_RB, cos_GB=cos_GB)
```
